데이터 설명

서울특별시 공동주택 아파트 정보는 서울 열린데이터 광장에 업로드 되어있고 서울특별시 공동주택통합마당 홈페이지에서 제공하는 아파트 정보입니다.
아파트명, 주소, 홈페이지, 연면적, 관리비부과면적, 세대타입, 난방방식, 건설사, 시행사, 준공일자 등 아파트관련 기본정보를 제공합니다.

저는 이 데이터를 통해서 위치나 교통등 주변환경에 의한 쾌적성이 아닌 아파트 자체 속성에 의한 쾌적성을 알아보고 싶습니다.
1. 전체 세대수 대비 주차대수를 구하여 주차 쾌적성을 알아봅니다.
2. 연면적 대비 전체세대수를 비교하여 주택 크기 의한 쾌적성을 알아보려고 합니다.
3. 각 자치구별 신축단지를 알아봅니다.
4. 각 자치구별 아파트 세대 규모를 알아봅니다.
5. 소셜믹스는 단지내 분양과 임대를 함께 조성하는 것입니다. 소셜믹스 그래프를 통해 전체 아파드 단지 대비 얼마나 소셜믹스가 진행중인지 알 수 있습니다.
