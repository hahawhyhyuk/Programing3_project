데이터 설명

서울특별시 공동주택 아파트 정보는 서울 열린데이터 광장에 업로드 되어있고 서울특별시 공동주택통합마당 홈페이지에서 제공하는 아파트 정보입니다.
아파트명, 주소, 홈페이지, 연면적, 관리비부과면적, 세대타입, 난방방식, 건설사, 시행사, 준공일자 등 아파트관련 기본정보를 제공합니다.

저는 이 데이터를 통해서 위치나 교통등 주변환경에 의한 쾌적성이 아닌 아파트 자체 속성에 의한 쾌적성을 알아보고 싶습니다.
전체 세대수 대비 주차대수를 구하여 주차 쾌적성을 알아보고, 연면적 대비 전체세대수를 구하여 밀도에 의한 쾌적성을 알아보려고 합니다.
입주일 대비 세대수를 비교하여 신축대단지를 알아보고, 각 자치구별 아파트 세대 규모를 알아봅니다.

열 설명

k-아파트코드 - 아파트 단지를 분류하는 고유 코드 입니다.
k-아파트명 - 아파트 단지 이름입니다.
k-단지분류(아파트,주상복합등등) - 아파트, 주상복합, 도시형생활주택, 연립주택으로 나뉩니다.
kapt도로명주소 - 아파트 단지의 도로명 주소입니다.
주소(시도)k-apt주소split - 아파트 단지의 시 입니다.
주소(시군구) - 아파트 단지의 소속 자치구 입니다.
주소(읍면동) - 아파트 단지의 소속 동입니다.
나머지주소 - 동이하의 상세주소입니다.
주소(도로명) - 아파트 단지의 도로 주소입니다.
주소(도로상세주소) - 도로명이하의 상세주소입니다.
k-전화번호 - 관리사무실의 전화번호입니다.
k-팩스번호 - 관리사무실의 팩스번호입니다.
단지소개기존clob - 
단지첨부파일 - 관리페이지에서의 첨부파일 입니다.
k-세대타입(분양형태) - 분양, 임대, 소셜믹스의 여부입니다.
k-관리방식 - 관리의 형태가 자치관리인지 위탁관리인지 입니다.
k-복도유형 - 복도의 형태가 계단식, 복도식, 혼합식, 타워식인지 여부입니다.
k-난방방식 - 세대별 난방 방식입니다.
k-전체동수 - 몇동을 건설했는지의 여부입니다.
k-전체세대수 - 전체 세대 수 입니다.
k-건설사(시공사) - 아파트의 건설사 입니다.
k-시행사 - 아파트 건설의 시행사입니다.
k-사용검사일-사용승인일 - 아파트의 사용 승인일 입니다.
k-연면적 - 아파트의 연면적입니다.
k-주거전용면적 - 연면적중 주거 전용 면적입니다.
k-관리비부과면적 - 관리비가 부과되는 면적입니다.
k-전용면적별세대현황(60㎡이하) - 60㎡이하의 세대수 입니다.
k-전용면적별세대현황(60㎡~85㎡이하) - 60㎡~85㎡이하의 세대수 입니다.
k-85㎡~135㎡이하 - 85㎡~135㎡이하의 세대수 입니다.
k-135㎡초과 - 135㎡초과의 세대수 입니다.
k-홈페이지 - 아파트 커뮤니티 홈페이지 주소입니다.
k-등록일자 - 서울특별시 공동주택 통합정보마당에 등록된 일자 입니다. 2017년 2월 이후만 등록되어있습니다.
k-수정일자 - 서울특별시 공동주택 통합정보마당에서 수정된 일자입니다.
고용보험관리번호 - 고용보험관리번호입니다.
경비비관리형태 - 경비의 위탁운영, 직영운영 여부입니다.
세대전기계약방법 - 전기의 종합계약 단일계약 여부입니다.
청소비관리형태 - 단지 청소비의 위탁운영, 직영운영 여부입니다.
건축면적 - 아파트의 건축면적입니다.
주차대수 - 단지의 주차대수 입니다.
기타/의무/임대/임의=1/2/3/4 - 
단지승인일 - 아파트 관리사무소가 공동주택 통합정보마당 홈페이지에 회원가입하고 서울시가 승인한 날짜입니다.
사용허가여부 - 건축물 사용허가여부입니다.
관리비 업로드 - 아파트 관리사무소가 공동주택 통합정보마당 홈페이지에 관리비 내역을 올렸는지의 여부입니다.
좌표X - x좌표값입니다.
좌표Y - y좌표값입니다.
단지신청일 - 단지신청일 입니다.
수정자 ID - 데이터의 수정자 ID입니다
